```python
# 1. Import Required Libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
```

```python
# 2. Load Dataset
df = pd.read_csv('Hospital_healthdata.csv.csv')
```

```python
# 3. Handle Missing Values
# Use median for numerical columns
numerical_cols = ['Age']
for col in numerical_cols:
    df[col] = df[col].fillna(df[col].median())
```

```python
# Use mode for categorical columns
categorical_cols = ['Difficulty Breathing', 'Blood Pressure', 'Cholesterol Level']
for col in categorical_cols:
    df[col] = df[col].fillna(df[col].mode()[0])
```

```python
# 4. Correct Invalid Ratings
df['Customer satisfaction mean'] = df['Customer satisfaction mean'].apply(lambda x: min(x, 5))
```

```python
# 5. Group Rare Diseases
rare_diseases = df['Disease'].value_counts()[df['Disease'].value_counts() < 10].index
df['Disease'] = df['Disease'].replace(rare_diseases, 'Special Disease')
```

```python
# 6. Encode Categorical Variables
binary_cols = ['Fever', 'Cough', 'Fatigue', 'Difficulty Breathing']
df[binary_cols] = df[binary_cols].replace({'Yes': 1, 'No': 0})
```

```python
# Encode ordinal variables
df['Blood Pressure'] = df['Blood Pressure'].replace({'Low': 0, 'Normal': 1, 'High': 2})
df['Cholesterol Level'] = df['Cholesterol Level'].replace({'Low': 0, 'Normal': 1, 'High': 2})
```

```python
# 7. Standardize Numerical Columns
scale_cols = ['Age', 'Customer satisfaction mean'] + df.select_dtypes(include='int64').columns.tolist()
scaler = StandardScaler()
df[scale_cols] = scaler.fit_transform(df[scale_cols])
```

```python
# 8. Visualizations
# Age distribution histogram
plt.figure(figsize=(8, 5))
sns.histplot(df['Age'], bins=30, kde=True)
plt.title('Age Distribution of Patients')
plt.xlabel('Standardized Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Gender distribution pie chart
plt.figure(figsize=(6, 6))
gender_counts = df['Gender'].value_counts()
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Gender Distribution of Patients')
plt.show()
```

```python
# Symptom frequency bar chart
plt.figure(figsize=(8, 5))
symptom_counts = df[binary_cols].sum()
sns.barplot(x=symptom_counts.index, y=symptom_counts.values)
plt.title('Frequency of Reported Symptoms')
plt.xlabel('Symptom')
plt.ylabel('Number of Patients')
plt.show()
```

```python
# Correlation heatmap
plt.figure(figsize=(12, 10))
corr = df.select_dtypes(include=['int64', 'float64']).corr()
sns.heatmap(corr, annot=False, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap of Numerical Variables')
plt.show()
```

```python
# 1. Classification Example: Predicting Disease
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report

# Prepare features and target
X_class = df.drop(columns=['Disease', 'Gender'])
y_class = df['Disease']

# Encode target labels
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y_class_encoded = le.fit_transform(y_class)

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X_class, y_class_encoded, test_size=0.2, random_state=42)

# Logistic Regression model
clf = LogisticRegression(max_iter=1000)
clf.fit(X_train, y_train)

# Prediction and evaluation
y_pred = clf.predict(X_test)
print("Classification Report (Disease Prediction):")
print(classification_report(y_test, y_pred, target_names=le.classes_))

```

```python
# 2. Clustering Example: Grouping Patients
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA

# Selecting relevant features for clustering
clustering_features = df[['Fever', 'Cough', 'Fatigue', 'Difficulty Breathing', 'Customer satisfaction mean']]

# Apply KMeans Clustering
kmeans = KMeans(n_clusters=3, random_state=42)
clusters = kmeans.fit_predict(clustering_features)
df['Cluster'] = clusters

# Visualize clusters using PCA
pca = PCA(n_components=2)
reduced_features = pca.fit_transform(clustering_features)

plt.figure(figsize=(8, 5))
sns.scatterplot(x=reduced_features[:, 0], y=reduced_features[:, 1], hue=clusters, palette='Set2')
plt.title('Patient Clusters Based on Symptoms and Satisfaction')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.show()
```

```python
# 3. Regression Example: Predicting Days in Hospital
import numpy as np
if 'Days in Hospital' not in df.columns:
    np.random.seed(42)
    df['Days in Hospital'] = np.random.randint(1, 15, size=len(df))

from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Define features and target
X_reg = df.drop(columns=['Days in Hospital', 'Gender', 'Disease', 'Cluster'])
y_reg = df['Days in Hospital']

# Train-test split
X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(X_reg, y_reg, test_size=0.2, random_state=42)

# Random Forest Regressor
regressor = RandomForestRegressor(n_estimators=100, random_state=42)
regressor.fit(X_train_reg, y_train_reg)

# Prediction and evaluation
y_pred_reg = regressor.predict(X_test_reg)
print("Regression Results (Days in Hospital Prediction):")
print("MAE:", mean_absolute_error(y_test_reg, y_pred_reg))
print("MSE:", mean_squared_error(y_test_reg, y_pred_reg))
print("R² Score:", r2_score(y_test_reg, y_pred_reg))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import classification_report, confusion_matrix, mean_absolute_error, mean_squared_error, r2_score

# 1. Logistic Regression: Satisfaction Classification
df['Satisfaction Category'] = df['Customer satisfaction mean'].apply(lambda x: 'High' if x <= 2 else 'Neutral' if x == 3 else 'Low')
le_sat = LabelEncoder()
y_satisfaction = le_sat.fit_transform(df['Satisfaction Category'])

X_satisfaction = df.drop(columns=['Customer satisfaction mean', 'Satisfaction Category', 'Gender', 'Disease', 'Days in Hospital', 'Cluster', 'Satisfaction_KNN'], errors='ignore')
X_satisfaction = X_satisfaction.select_dtypes(include=[np.number])

X_train_sat, X_test_sat, y_train_sat, y_test_sat = train_test_split(X_satisfaction, y_satisfaction, test_size=0.2, random_state=42)

logreg = LogisticRegression(max_iter=1000)
logreg.fit(X_train_sat, y_train_sat)
y_pred_sat = logreg.predict(X_test_sat)

print("Logistic Regression Classification Report:")
print(classification_report(y_test_sat, y_pred_sat, target_names=le_sat.classes_))

# Confusion Matrix
cm = confusion_matrix(y_test_sat, y_pred_sat)
plt.figure(figsize=(6, 4))
sns.heatmap(cm, annot=True, fmt='d', xticklabels=le_sat.classes_, yticklabels=le_sat.classes_, cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix - Satisfaction Classification')
plt.show()

```

```python
# 2. Decision Tree: Disease Prediction
le_disease = LabelEncoder()
df['Disease_encoded'] = le_disease.fit_transform(df['Disease'])
y_disease = df['Disease_encoded']
X_disease = df.drop(columns=['Disease', 'Disease_encoded', 'Gender', 'Satisfaction Category', 'Days in Hospital', 'Cluster', 'Satisfaction_KNN'], errors='ignore')
X_disease = X_disease.select_dtypes(include=[np.number])

X_train_dis, X_test_dis, y_train_dis, y_test_dis = train_test_split(X_disease, y_disease, test_size=0.2, random_state=42)
tree = DecisionTreeClassifier(random_state=42)
tree.fit(X_train_dis, y_train_dis)
y_pred_dis = tree.predict(X_test_dis)

print("Decision Tree Classification Report:")
print(classification_report(y_test_dis, y_pred_dis, target_names=le_disease.classes_))

# Feature Importance
importances = tree.feature_importances_
features = X_disease.columns
indices = np.argsort(importances)[-10:]

plt.figure(figsize=(8, 6))
plt.barh(range(len(indices)), importances[indices], align='center')
plt.yticks(range(len(indices)), [features[i] for i in indices])
plt.xlabel('Importance')
plt.title('Top 10 Feature Importances - Disease Prediction')
plt.show()

```

```python
# 3. KNN: Gender-Based Satisfaction
df['Gender_num'] = df['Gender'].map({'Male': 0, 'Female': 1})
df['Satisfaction_KNN'] = df['Customer satisfaction mean'].apply(lambda x: 'High' if x <= 2 else 'Neutral' if x == 3 else 'Low')
le_knn = LabelEncoder()
y_knn = le_knn.fit_transform(df['Satisfaction_KNN'])

X_knn = df[['Gender_num']]
X_train_knn, X_test_knn, y_train_knn, y_test_knn = train_test_split(X_knn, y_knn, test_size=0.2, random_state=42)
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train_knn, y_train_knn)
y_pred_knn = knn.predict(X_test_knn)

print("KNN Classification Report:")
print(classification_report(y_test_knn, y_pred_knn, target_names=le_knn.classes_))

# Bar plot: Satisfaction by Gender
plt.figure(figsize=(7, 5))
sns.countplot(data=df, x='Satisfaction_KNN', hue='Gender')
plt.title('Satisfaction Category by Gender')
plt.xlabel('Satisfaction Level')
plt.ylabel('Count')
plt.show()

```

```python
# 4. Random Forest: Regression
if 'Days in Hospital' not in df.columns:
    np.random.seed(42)
    df['Days in Hospital'] = np.random.randint(1, 15, size=len(df))

X_reg = df.drop(columns=['Days in Hospital', 'Gender', 'Disease', 'Satisfaction Category', 'Cluster', 'Satisfaction_KNN'], errors='ignore')
X_reg = X_reg.select_dtypes(include=[np.number])
y_reg = df['Days in Hospital']

X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(X_reg, y_reg, test_size=0.2, random_state=42)
rfr = RandomForestRegressor(n_estimators=100, random_state=42)
rfr.fit(X_train_reg, y_train_reg)
y_pred_reg = rfr.predict(X_test_reg)

print("Random Forest Regression:")
print("MAE:", mean_absolute_error(y_test_reg, y_pred_reg))
print("MSE:", mean_squared_error(y_test_reg, y_pred_reg))
print("R² Score:", r2_score(y_test_reg, y_pred_reg))

# Actual vs Predicted
plt.figure(figsize=(8, 5))
sns.scatterplot(x=y_test_reg, y=y_pred_reg)
plt.plot([y_test_reg.min(), y_test_reg.max()], [y_test_reg.min(), y_test_reg.max()], 'r--')
plt.xlabel('Actual Days in Hospital')
plt.ylabel('Predicted Days in Hospital')
plt.title('Actual vs. Predicted Days in Hospital')
plt.grid(True)
plt.show()

# Residual Distribution
residuals = y_test_reg - y_pred_reg
plt.figure(figsize=(8, 5))
sns.histplot(residuals, kde=True, bins=20)
plt.title('Residual Distribution')
plt.xlabel('Residual (Actual - Predicted)')
plt.grid(True)
plt.show()



```
